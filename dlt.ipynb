```python
import google.generativeai as genai

from dotenv import load_dotenv
import os

load_dotenv() # load env variables into system
api_key = os.getenv("GEMINI_API_KEY") # get from system

try:
    model = genai.GenerativeModel("gemini-2.5-flash")
    response = model.generate_content("How do i list mdoels in my google genai free tier api key")

    print(response.text)

except Exception as e:
    print("ERROR:", e)
```
