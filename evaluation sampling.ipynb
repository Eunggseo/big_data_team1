New version 

```python
import pandas as pd

input_file = "discharge.csv"
output_file = "sampled_notes_20.csv"
chunk_size = 10000

sample_rows = []

try:
    chunks = pd.read_csv(input_file, chunksize=chunk_size)
    
    for chunk in chunks:
        sample_rows.append(chunk.sample(n=1, random_state=42))

    df_combined = pd.concat(sample_rows)
    df_sampled = df_combined.sample(n=20, random_state=42).reset_index(drop=True)

    result_df = pd.DataFrame({
        'id': range(1, 21), 
        'note_id': df_sampled['note_id'],
        'text': df_sampled['text']
    })

    result_df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"Successfully saved the 20 sampled results to: {output_file}\n")

    for _, row in result_df.iterrows():
        print(f"--- Note {row['id']} (note_id: {row['note_id']}) ---")
        print(f"{str(row['text'])[:1000]}\n")
        print("-" * 50)

except FileNotFoundError:
    print(f"Error: File '{input_file}' not found. Please check if the file path is correct.")
except Exception as e:
    print(f"An error occurred during execution: {e}")
```
